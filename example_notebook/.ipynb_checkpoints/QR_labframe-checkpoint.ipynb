# Pulse shape definition for Time-dependent simulation
---

```python
def MWshape(t, args): # simple ideal square pulse
    amp = args['amp']
    phas = args['phase']
    f = args['freq']
    return amp * np.cos(f*t + phas)

def MWshape_SSBed(t, args): # ssb pulse # sampling rate = 1Gsa/s # No filtering
    I = args['I']
    Q = args['Q']
    f = args['freq']
    j = int(t)
    i, q = I[j], Q[j]
    amp = np.sqrt(i**2 + q**2)
    phi = np.arctan2(i, q)
    pulse = amp * np.cos(f*t + phi)
    return pulse

def MWshape_SSB_DRAGid(t, args): # ssb pulse # sampling rate = 1Gsa/s # No filtering
    mwamp = args['amp']
    Tg = args['Tg']
    sigma = args['sigma']
    f = args['freq']
    i = gaussianDRAG(Tg, sigma, t=t) * mwamp
    q = dGaussianDRAG(Tg, sigma, t=t) * mwamp
    amp = np.sqrt(i**2 + q**2)
    phi = np.arctan2(i, q)
    pulse = amp * np.cos(f*t + phi)
    return pulse
```

# Simple pulse test
---

```python
Hlab = cqed.Hlab
B = cqed.B
X = cqed.X
Y = cqed.Y
Z = cqed.Z
Num = cqed.Na
A = cqed.A
S0 = scon.iniState1Q1Rsys(Nq, Nf, 0, 0, mode='rho')
S1 = scon.iniState1Q1Rsys(Nq, Nf, 1, 0, mode='rho')
S2 = scon.iniState1Q1Rsys(Nq, Nf, 2, 0, mode='rho')
e_ops1 = [X, Y, Z]
e_ops2 = [S0, S1, S2]
Hd_real = cqed.Hd1_real
Hd_imag = cqed.Hd1_imag

psi0 = scon.iniState1Q1Rsys(Nq, Nf, 0, 0, mode='ket')
```

## See qutip document
---

```python
Tg = 150 # unit is [ns]
t_list = np.linspace(0, Tg, int(Tg+1))
H = [Hlab, [Hd_real, MWshape]]
args = {'amp':0.5, 'phase':np.pi*0., 'freq':fq1}

# output = qt.mesolve(H, psi0, t_list, args=args, c_ops=[], \
#                     e_ops=[cqed.X, cqed.Y, cqed.Z], options=qt.Options(nsteps=1e4))
output = qt.mesolve(H, psi0, t_list, args=args, c_ops=[], options=qt.Options(nsteps=1e4))
Urots = cqed.calcUrot(t_list, fq1) # Unitary operator for frame transformation
psi_lab = output.states # lab-frame states
psi_rot = [Urots[i]*psi_lab[i] for i in range(len(t_list))] # Constructing rotating frame states
expects = qt.expect([X, Y, Z, Num], psi_rot)
```

```python
plt.figure('expect', dpi=100)
plt.xlabel('time [ns]')
plt.ylabel('Expectation value')
plt.plot(t_list, expects[0], label=r'$\langle X \rangle_{transmon}$')
plt.plot(t_list, expects[1], label=r'$\langle Y \rangle_{transmon}$')
plt.plot(t_list, expects[2], label=r'$\langle Z \rangle_{transmon}$')
plt.legend()
```

```python
def bloch(evals):
    # ブロッホ球を描画するだけ
    b = qt.Bloch()
    
    b.add_vectors([evals[0][0], evals[1][0], evals[2][0]])
    b.add_vectors([evals[0][-1], evals[1][-1], evals[2][-1]])
    b.add_points(evals, meth='l')
    b.make_sphere()

    b.show()

bloch(expects)
```
